```python
# 分割 split()
text = "苹果,香蕉,橘子,葡萄"
fruits = text.split(",")
print(f"分割结果：{fruits}")
print(f"第一个水果：{fruits[0]}")
url = "https://www.lanqiao.cn/users/3820686/?tab=1&lq_cup_tab=1"
parts = url.split("?")
print(f"URL路径：{parts[0]}")
print(f"参数部分：{parts[1]}")
params = "name=张三&age=20&city=重庆"
pairs = params.split("&")
for pair in pairs:
    key, value = pair.split("=")
    print(f"{key}：{value}")
```

```python
# 大小写转换
text = "Hello PyThon WorLd"
print(f"原字符串：{text}")
print(f"upper()：{text.upper()}")
print(f"lower()：{text.lower()}")
print(f"capitalize()：{text.capitalize()}")
print(f"title()：{text.title()}")
```

```python
# 去除空格
# strip()：去除两端空格  lstrip()：去除左侧空格     rstrip()：去除右侧空格
s = "   Hello Python   "
print(f"原字符串：'{s}'")
print(f"strip()：'{s.strip()}'")
print(f"lstrip()：'{s.lstrip()}'")
print(f"rstrip()：'{s.rstrip()}'")
```

```python
# 查找
s = "一闪一闪亮晶晶，满天都是小星星"
p1 = s.index("一闪")    # 找到了返回一闪的位置, 找不到报错
p2 = s.find("闪", 2)  # 从索引2开始找
p3 = s.find("大星星")    # 找不到返回-1
print(f"原字符串：{s}")
print(f"一闪的位置：{p1}")
print(f"闪的位置：{p2}")
print(f"大星星的位置，找不到返回：{p3}")
```

```python
# 连接 join()
# 语法：分隔符.join(可迭代对象)    将列表/元组中的元素用指定分隔符连接
words = ["苹果", "香蕉", "橘子"]
result = "-".join(words)    # 以'-'为连接符连接
print(f"连接结果：{result}")
items = ["A", "B", "C"]
print(f"逗号分隔：{','.join(items)}")
print(f"空格分隔：{' '.join(items)}")
print(f"空字符串：{''.join(items)}")
```

```python
# 替换 replace()
# 语法：replace(旧子串, 新子串)  返回新字符串，原字符串不变
text = "我喜欢苹果，苹果很好吃"
new_text = text.replace("苹果", "香蕉")
print(f"原字符串：{text}")
print(f"替换后：{new_text}")
s = "aaa bbb aaa ccc aaa"
result = s.replace("aaa", "xxx", 2)  # 只替换前2个
print(f"原字符串：{s}")
print(f"替换2次：{result}")
```
