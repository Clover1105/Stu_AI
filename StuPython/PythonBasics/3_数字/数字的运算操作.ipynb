```python
# 算术运算
print("============= 算术运算 =============")
a = 10
b = 3
print(f"a = {a}, b = {b}")
print(f"a + b = {a + b}")   # 加法
print(f"a - b = {a - b}")   # 减法
print(f"a * b = {a * b}")   # 乘法
print(f"a / b = {a / b}")   # 除法（结果是浮点数）
print(f"a ** b = {a ** b}") # 幂运算：10的3次方
print(f"a % b = {a % b}")   # 取余：10除以3余1
print(f"a // b = {a // b}") # 整除：10除以3得3
```

```python
# 比较运算
print("============= 比较运算 =============")
x = 15
y = 20
print(f"x = {x}, y = {y}")
print(f"x > y  : {x > y}")     # 大于
print(f"x < y  : {x < y}")     # 小于
print(f"x >= y : {x >= y}")    # 大于等于
print(f"x <= y : {x <= y}")    # 小于等于
print(f"x == y : {x == y}")    # 等于
print(f"x != y : {x != y}")    # 不等于
print(f"x is y : {x is y}")    # 判断是否是同一个对象
print(f"x is not y : {x is not y}")  # 判断是否不是同一个对象
```

```python
# 逻辑运算
print("============= 逻辑运算 =============")
p = True
q = False
print(f"p = {p}, q = {q}")
print(f"p and q : {p and q}")   # 与：两个都为True才为True
print(f"p or q  : {p or q}")    # 或：至少一个为True就为True
print(f"not p   : {not p}")     # 非：取反
print("--- 短路特性演示 ---")
print(f"10 and 20 : {10 and 20}")      # 左边为真，返回右边
print(f"0 and 100 : {0 and 100}")      # 左边为假，返回左边
print(f"10 or 0   : {10 or 0}")        # 左边为真，返回左边
print(f"0 or 20   : {0 or 20}")        # 左边为假，返回右边
'''
短路总结：x = A and B
A运行的结果转化为布尔值 -->
    如果为真，B运行，x的结果就是B运行的结果
    如果为假，B不运行，x的结果就是A运行的结果
'''
# 实际应用：判断分数是否合格
print("--- 应用题 ---")
score = 85
is_pass = score >= 60 and score <= 100
print(f"分数{score}是否合格：{is_pass}")
```

```python
# 赋值运算
print("============= 赋值运算 =============")
num = 10
print(f"初始值: num = {num}")
num += 5   # 等价于 num = num + 5
print(f"num += 5 --> {num}")
num -= 3   # 等价于 num = num - 3
print(f"num -= 3 --> {num}")
num *= 2   # 等价于 num = num * 2
print(f"num *= 2 --> {num}")
num /= 4   # 等价于 num = num / 4
print(f"num /= 4 --> {num}")
num = 20
num //= 3  # 等价于 num = num // 3
print(f"num //= 3 --> {num}")
num = 17
num %= 5   # 等价于 num = num % 5
print(f"num %= 5 --> {num}")
num = 2
num **= 3  # 等价于 num = num ** 3
print(f"num **= 3 --> {num}")
```
