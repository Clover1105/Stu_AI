```python
"""
字典其他方法
    - update()：合并字典
    - setdefault()：键不存在时设置默认值，存在时不修改
    - copy()：浅拷贝
    - in：判断键是否存在
"""
```

```python
# update()合并字典
dict1 = {"a": 1, "b": 2}
dict2 = {"b": 3, "c": 4, "d": 5}
dict1.update(dict2)  # 重复的键会被覆盖
print(f"合并后：{dict1}")
```

```python
# setdefault()设置默认值
person = {"name": "小明"}
person.setdefault("age", 18)  # age不存在，设置为18
person.setdefault("name", "小红")  # name存在，不修改
print(f"setdefault后：{person}")
```

```python
# copy()浅拷贝
stu = {"name": "张三", "scores": [90, 85]}
copy_stu = stu.copy()
copy_stu["scores"][0] = 100  # 修改拷贝中的可变对象会影响原字典
print(f"原字典：{stu}")
print(f"拷贝字典：{copy_stu}")
```

```python
# in判断键是否存在
stu = {"name": "张三", "scores": [90, 85]}
print(f"'name'在字典中：{'name' in stu}")
print(f"'age'在字典中：{'age' in stu}")
```
