```python
"""
列表常用方法
    - append()：末尾添加
    - pop()：删除并返回元素
    - remove()：删除指定元素
    - insert()：指定位置插入
    - del：删除
"""
# append(): 末尾添加
num = [1,2,3,4,5,6,7,8,9]
num.append(10)
print(f"append后：{num}")

# pop(): 删除并返回元素
l = num.pop(7)  # 移除指定下标的元素，并返回该元素，默认删除最后一个元素
print(f"pop后：{num}, 返回的元素是：{l}")

# remove(): 删除指定元素
l = num.remove(5)   # 移除指定元素，并返回None
print(f"remove后：{num}, 返回的是：{l}")

# insert(): 指定位置插入
num.insert(2, 5)    # 把5插入到索引为2的位置 以前的2后面的元素都向后移动一位
print(f"insert后：{num}")

# del关键字
# del num # 删除变量, 变量名不存在会报错
del num[6]
print(f"del后：{num}")
```

```python
"""
列表其他常用方法
    - max()/min()：最大/最小值
    - count()：统计出现次数
    - extend()：扩展列表
    - reverse()：反转列表
    - clear()：清空列表
"""
# max()/min()：最大/最小值
scores = [85, 92, 78, 90, 88]
print(f"最高分：{max(scores)}")
print(f"最低分：{min(scores)}")

# count()：统计出现次数
numbers = [1, 2, 2, 3, 2, 4]
print(f"2出现了{numbers.count(2)}次")

# extend()：扩展列表
l1 = [1, 3, 2]
l2 = [5, 4]
l1.extend(l2)
print(f"扩展后：{l1}")

# reverse()：反转列表
l1.reverse()
print(f"反转后：{l1}")

# clear()：清空列表
l1.clear()
print(f"清空后：{l1}")
```
