```python
# 1. 编写一个函数 reverse_string(s) ，接受字符串，返回反转后的字符串（不使用切片）。
def reverse_string(s):
    s1 = ""
    for i in s:
        s1 = i + s1
    return s1
a = "hello world"
print(reverse_string(a))
```

```python
# 2. 编写一个函数 is_palindrome(n) ，判断一个正整数是否是回文数。
def is_palindrome(n):
    n=str(n)
    n1=n[::-1]
    if n==n1:
        return True
    else:
        return False
print(is_palindrome(12321))
```

```python
# 3. 编写一个函数 flatten_list(nested) ，将嵌套列表展开为单层列表。
# def flatten_list(nested):
#     ls = []
#     for i in nested:
#         try:
#             ls.extend(flatten_list(i))
#         except TypeError:
#             ls.append(i)
#     return ls
# l = [1,[2,3],[[4,5]]]
# print(flatten_list(l))

# 老师的代码
def flatten_list(nested):
    if not isinstance(nested,list):
        return [nested]
    arr=[]
    for el in nested:
        el2=flatten_list(el)
        arr+=el2
    return arr
l = [1,[2,3],[[4,5]]]
print(flatten_list(l))
```

```python
# 4. 编写一个函数 count_words(text) ，统计字符串中每个单词出现的次数，返回字典。
def count_words(text):
    ls = text.split()
    d = {}
    for i in ls:
        d[i] = ls.count(i)
    return d
s = "hello world hello world hello"
print(count_words(s))

# 老师的代码
def count_words(text):
    arr = text.split()
    return {el:arr.count(el) for el in arr}
s = "hello world hello world hello"
print(count_words(s))
```

```python
# 7. 编写一个函数 memoize(f) ，实现记忆化装饰器，缓存函数结果。
# 老师的代码
def memoize(cb):
    arr = []
    def f():
        res = cb()
        arr.append(res)
        return arr
    return f
@memoize
def test():
    return 10
print(test())
print(test())
print(test())
```

```python
# 8. 编写一个生成器函数 primes(n) ，生成前n个素数(大于1)。
# def primes(n):
#     for i in range(2,n+1):
#         for j in range(2,i):
#             if i%j==0:
#                 break
#         else:
#             yield i
# p = primes(10)
# try:
#     for i in range(10):
#         print(next(p))
# except StopIteration:
#     print("end")

# 老师的代码
def is_prime(n):
    if n == (1 or 2):
        return True
    for i in range(2,n//2+1):
        if n % i == 0:
            return False
    return True
def primes(n):
    count = 1
    start = 1
    while count <= n:
        if is_prime(start):
            count += 1
            yield start
        start += 1
res = primes(5)
for i in range(5):
    print(next(res))

```

```python
# 9. 编写一个类 Stack ，实现栈的 push、pop、peek、is_empty 操作。
class Stack:
    def __init__(self):
        self.ls = []
    def push(self,x):   # 添加
        self.ls.append(x)
    def pop(self):  # 删除
        return self.ls.pop()
    def peek(self): # 查看栈顶元素
        return self.ls[-1]
    def is_empty(self): # 判断是否为空
        return len(self.ls) == 0
s = Stack() # 创建一个栈对象，同时调用__init__()为该对象创建一个空列表
s.push(1)
s.push(2)
s.push(3)
print(s.ls)
print(s.pop())
print(s.peek())
print(s.is_empty())
```

```python
# 10. 编写一个类 Student ：
# 属性：name, age, scores（字典）
# 方法：add_score(subject, score)、get_average()、get_letter_grade()
# 字母等级：A(90+), B(80+), C(70+), D(60+), F(<60)
class Student:
    # def __init__(self,name,age,scores):
    def __init__(self,name,age):
        self.name = name
        self.age = age
        # self.scores = scores
        self.scores = {}
    def add_score(self,subject,score):  # 添加成绩
        self.scores[subject] = score
    def get_average(self):  # 计算平均分
        # avg = sum(self.scores.values()) / len(self.scores)
        all_score = list(self.scores.values())
        average = sum(all_score) / len(all_score)
        print(f"{self.name}的平均分为{average:.2f}")
        return average
    def get_letter_grade(self): # 计算等级
        avg = self.get_average()
        if avg >= 90:
            return "A"
        elif avg >= 80:
            return "B"
        elif avg >= 70:
            return "C"
        elif avg >= 60:
            return "D"
        else:
            return "F"
s = Student("Tom", 20)
print(s.name, s.age)
s.add_score("linux", 90)
s.add_score("python", 80)
s.add_score("java", 70)
print(s)
print(s.scores)
# print(s.get_average())
print(s.get_letter_grade())
```
