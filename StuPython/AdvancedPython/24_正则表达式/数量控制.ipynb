```python
"""
基本量词
    - *：重复0次或更多次
    - +：重复1次或更多次
    - ?：重复0次或1次
"""

import re

text = "华清远见 华清666远见 华清6远见 华清66远见"

# * 匹配0次或多次
data1 = re.findall("华清6*远见", text)
print(f"6* ：{data1}")

# + 匹配1次或多次
data2 = re.findall("华清6+远见", text)
print(f"6+ ：{data2}")

# ? 匹配0次或1次
data3 = re.findall("华清6?远见", text)
print(f"6? ：{data3}")
```

```python
"""
花括号
    - {n}：重复n次
    - {n,}：重复n次或更多次
    - {n,m}：重复n到m次
"""

import re

text = "华清远见 华清666远见 华清6远见 华清66远见"

# {3} 匹配恰好3次
data1 = re.findall("华清6{3}远见", text)
print(f"6{{3}} ：{data1}")

# {2,} 匹配2次或更多
data2 = re.findall("华清6{2,}远见", text)
print(f"6{{2,}}：{data2}")

# {0,2} 匹配0到2次
data3 = re.findall("华清6{0,2}远见", text)
print(f"6{{0,2}}：{data3}")
```
