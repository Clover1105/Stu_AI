```python
"""
findall基本使用
    - re.findall(规则, 文本) 查找所有匹配的内容
    - 直接匹配文本中的指定字符串
"""

import re

text = "hqyj牛皮6666,hqyj有个老师也牛皮666"
data = re.findall("hqyj", text)
print(f"匹配'hqyj'：{data}")  # ['hqyj', 'hqyj']
```

```python
"""
字符集
    - [abc]：匹配a或b或c任意一个字符
    - [^abc]：匹配除了a、b、c以外的任意字符
    - [a-z]：匹配a到z的任意字符
"""

import re

text = "hqyj牛皮6666,hqyj有个老师也牛皮666"

# [hqyj] 匹配h或q或y或j
data1 = re.findall("[hqyj]", text)
print(f"[hqyj]：{data1}")

# [^hqyj] 匹配除hqyj外的字符
data2 = re.findall("[^hqyj]", text)
print(f"[^hqyj]：{data2}")

# [a-z]hqyj 匹配a-z后跟hqyj
data3 = re.findall("[a-z]hqyj", text)
print(f"[a-z]hqyj：{data3}")
```

```python
"""
通配符点(.)
    - . 匹配除了换行符以外的任意单个字符
    - .+ 贪婪匹配：匹配尽可能长的内容
    - .? 非贪婪匹配：匹配尽可能短的内容
"""

import re

text = "hqyj牛皮6666,hqyj有个老师abchqyj也牛皮666"

# .hqyj 匹配任意一个字符后跟hqyj
data1 = re.findall(".hqyj", text)
print(f".hqyj：{data1}")  # [',hqyj', 'chqyj']

# .+hqyj 贪婪匹配：匹配最长的
data2 = re.findall(".+hqyj", text)
print(f".+hqyj（贪婪）：{data2}")

# .?hqyj 非贪婪匹配：匹配最短的
data3 = re.findall(".?hqyj", text)
print(f".?hqyj（非贪婪）：{data3}")
```

```python
"""
预定义字符集
    - \w：匹配字母、数字、下划线、汉字
    - \d：匹配数字
    - \s：匹配空白字符（空格、制表符等）
"""

import re

# \w 匹配字母数字下划线汉字
text1 = "华清_远见abc 华清hqyj远见"
data1 = re.findall("华清\w+远见", text1)
print(f"\\w匹配：{data1}")

# \d 匹配数字
text2 = "hqyj66d6 a1h43d3fd43s43d4"
data2 = re.findall("d\\d+", text2)
print(f"\\d匹配：{data2}")

# \s 匹配空白符
text3 = "hqyj666  jack karen 666"
data3 = re.findall("\\sj\\w+\\s", text3)
print(f"\\s匹配：{data3}")
```
