```python
"""
re.match方法
    - 从字符串起始位置匹配
    - 成功返回匹配对象，失败返回None
    - 常用方法：group()、start()、end()、span()
"""

import re

text = "www.python.com"

# 起始位置匹配
result1 = re.match('www', text)
print(f"匹配结果：{result1}")
print(f"匹配内容：{result1.group()}")
print(f"开始位置：{result1.start()}")
print(f"结束位置：{result1.end()}")
print(f"位置元组：{result1.span()}")

# 不在起始位置匹配
result2 = re.match('com', text)
print(f"\n不在开头匹配：{result2}")
```

```python
"""
re.search方法
    - 扫描整个字符串，返回第一个成功匹配
    - 成功返回匹配对象，失败返回None
"""

import re

text = "www.hqyj.com"

# 从开头匹配
result1 = re.search('www', text)
print(f"开头匹配：{result1.span()}")

# 从中间匹配
result2 = re.search('com', text)
print(f"中间匹配：{result2.span()}")
```

```python
"""
re.sub替换方法
    - 替换匹配成功的字符
    - 类似字符串的replace方法
    - 语法：re.sub(规则, 替换内容, 文本)
"""

import re

text = "以前华清远见在四川大学旁边,现在华清远见在西南交大旁边"

data = re.sub("华清远见", "北京华清远见科技集团成都中心", text)
print(f"替换前：{text}")
print(f"替换后：{data}")
```

```python
"""
re.split分割方法
    - 根据匹配的位置对字符串进行分割
    - 返回分割后的列表
"""

import re

text = "python is   very easy"

# 根据一个或多个空格分割
data = re.split("\\s{1,}", text)
print(f"原字符串：{text}")
print(f"分割结果：{data}")
```

```python
"""
re.finditer迭代方法
    - 类似findall但返回迭代器
    - 匹配大量数据时节省内存
    - 每个元素是匹配对象
"""

import re

text = "python is   very easy"

# findall返回列表
data1 = re.findall("\\w+", text)
print(f"findall返回：{data1}")

# finditer返回迭代器
data2 = re.finditer("\\w+", text)
print(f"finditer返回：{data2}")

# 遍历迭代器获取结果
print("遍历结果：")
for el in data2:
    print(f"  {el.group()}")
```
