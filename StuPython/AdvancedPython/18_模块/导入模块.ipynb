```python
"""
import导入模块
    - 使用import 模块名导入整个模块
    - 使用时需要模块名.标识符的形式访问
"""

# 导入math模块并使用其功能
import math

print(f"圆周率：{math.pi}")
print(f"平方根：{math.sqrt(16)}")
print(f"5的阶乘：{math.factorial(5)}")
```

```python
"""
from...import导入
    - from 模块名 import 标识符1, 标识符2
    - 直接使用标识符名，不需要模块名前缀
"""

# 题目：从math模块中导入特定函数和变量
from math import pi, sqrt, factorial

print(f"圆周率：{pi}")
print(f"平方根：{sqrt(25)}")
print(f"5的阶乘：{factorial(5)}")
```

```python
"""
from...import *
    - from 模块名 import * 导入模块中所有内容
    - 使用时直接使用标识符名
"""

# 使用import *导入math模块
from math import *

print(f"圆周率：{pi}")
print(f"自然常数e：{e}")
print(f"平方根：{sqrt(100)}")
print(f"正弦值：{sin(pi/2)}")
```

```python
"""
导入并取别名（as）
    - import 模块名 as 别名：给模块取别名
    - from 模块名 import 标识符 as 别名：给标识符取别名
"""

# 给模块取别名
import math as m

print(f"使用别名访问pi：{m.pi}")
print(f"使用别名访问sqrt：{m.sqrt(49)}")


# 给标识符取别名
from datetime import datetime as dt

now = dt.now()
print(f"当前时间：{now}")
```
