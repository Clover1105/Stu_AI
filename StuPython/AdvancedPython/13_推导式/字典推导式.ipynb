```python
"""
字典推导式基础
    - 语法：{键表达式: 值表达式 for 变量 in 可迭代对象}
    - 快速生成字典
    - 键必须是不可变类型
"""
# 1.将列表转为字典（元素作为键，索引作为值）
sg_s = ["苹果", "香蕉", "橘子", "葡萄"]
sgs_dict = {sg: idx for idx, sg in enumerate(sg_s)}
print(f"水果字典：{sgs_dict}")

# 2.生成数字-平方字典
squ = {x: x**2 for x in range(1, 8)}
print(f"1-7的平方字典：{squ}")

# 3.创建字母-ASCII码字典
ls = ['a', 'b', 'c', 'd', 'e']
ascii_dict = {ch: ord(ch) for ch in ls}
print(f"字母ASCII码：{ascii_dict}")

# 4.生成数字-奇偶性字典
ji_ou = {n: "奇数" if n % 2 else "偶数" for n in range(1, 11)}
print(f"数字奇偶性：{ji_ou}")
```

```python
"""
带条件的字典推导式
    - 语法：{键: 值 for 变量 in 可迭代对象 if 条件}
    - 只包含满足条件的键值对
    - 可以结合if-else进行值处理
"""
# 1.只保留成绩大于80的学生
scores = {"张三": 85, "李四": 72, "王五": 93, "赵六": 68, "小明": 88}
high_scores = {name: score for name, score in scores.items() if score >= 80}
print(f"高分学生（>=80）：{high_scores}")

# 2.筛选出年龄大于等于18的人
people = {"张三": 16, "李四": 25, "王五": 18, "赵六": 14, "小明": 30}
adults = {name: age for name, age in people.items() if age >= 18}
print(f"成年人：{adults}")

# 3.筛选出长度大于3的单词及其长度
words = ["cat", "elephant", "dog", "tiger", "bird"]
word_lengths = {word: len(word) for word in words if len(word) > 3}
print(f"长度>3的单词及长度：{word_lengths}")

# 4.筛选出值为正数的键值对
data = {"a": 10, "b": -5, "c": 20, "d": -8, "e": 15}
positive = {k: v for k, v in data.items() if v > 0}
print(f"正数键值对：{positive}")
```

```python
"""
字典键值交换
    - 使用字典推导式交换键和值
    - 注意：值必须可哈希（不可变）
    - 值重复时后面的会覆盖前面的
"""
# 1.基本键值交换
people = {"name": "张三", "age": 18, "city": "北京"}
jh = {v: k for k, v in people.items()}
print(f"原字典：{people}")
print(f"键值交换后：{jh}")

# 2.学生信息交换（学号-姓名）
students = {"S001": "小明", "S002": "小红", "S003": "小刚"}
reverse_students = {name: id for id, name in students.items()}
print(f"原字典：{students}")
print(f"交换后（姓名-学号）：{reverse_students}")

# 3.成绩等级交换
grades = {"A": 90, "B": 80, "C": 70, "D": 60}
score_to_grade = {score: grade for grade, score in grades.items()}
print(f"分数对应等级：{score_to_grade}")

# 4.注意值重复的情况
chong_fu = {"a": 1, "b": 2, "c": 1}
jh = {v: k for k, v in chong_fu.items()}
print(f"值重复的字典：{chong_fu}")
print(f"交换后（1被覆盖）：{jh}")
```
