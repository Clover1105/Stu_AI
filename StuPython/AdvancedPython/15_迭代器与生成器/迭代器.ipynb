```python
"""
创建迭代器
    - 使用iter()函数将可迭代对象转换为迭代器
    - 迭代器是一个可以记住遍历位置的对象
    - 列表、字符串、元组等都是可迭代对象
"""
# 1.将列表转换为迭代器
fruits = ["苹果", "香蕉", "橘子", "葡萄"]
fruit_iter = iter(fruits)
print(f"迭代器对象：{fruit_iter}")
print(f"迭代器类型：{type(fruit_iter)}")  # <class 'list_iterator'>

# 2.将字符串转换为迭代器
text = "Python"
text_iter = iter(text)
print(f"字符串迭代器：{text_iter}")

# 3.将range对象转换为迭代器
num_range = range(1, 6)
range_iter = iter(num_range)
print(f"range迭代器：{range_iter}")
```

```python
"""
next()获取迭代器元素
    - 使用next()函数从迭代器中逐个取出元素
    - 每次调用next()返回下一个元素
    - 迭代器只能往前取，不能后退
"""
# 1.逐个获取迭代器中的元素
numbers = [10, 20, 30, 40]
num_iter = iter(numbers)
print("逐个取出元素：")
r1 = next(num_iter)
print(f"第1次取：{r1}")
r2 = next(num_iter)
print(f"第2次取：{r2}")
r3 = next(num_iter)
print(f"第3次取：{r3}")
r4 = next(num_iter)
print(f"第4次取：{r4}")


# 2.连续调用next()
colors = ["红", "黄", "蓝"]
color_iter = iter(colors)
print(f"取出：{next(color_iter)}")
print(f"取出：{next(color_iter)}")
print(f"取出：{next(color_iter)}")
```

```python
"""
迭代器耗尽与StopIteration
    - 当迭代器中没有元素时，继续调用next()会抛出StopIteration异常
    - 这是迭代器结束的信号
    - 可以用try-except捕获异常
"""
# 1.迭代器耗尽后继续取会报错
fruits = ["苹果", "香蕉"]
fruit_iter = iter(fruits)
print(next(fruit_iter))  # 苹果
print(next(fruit_iter))  # 香蕉
# print(next(fruit_iter))  # 取消注释会报错 StopIteration


# 2.使用try-except捕获StopIteration
numbers = [1, 2]
num_iter = iter(numbers)
try:
    print(next(num_iter))
    print(next(num_iter))
    print(next(num_iter))  # 这里会触发异常
except StopIteration:
    print("迭代器已耗尽，没有更多元素了")


# 3.验证迭代器状态
data = [100]
data_iter = iter(data)
print(f"第一次取：{next(data_iter)}")
print("迭代器现在已空")
```

```python
"""
for循环遍历迭代器
    - for循环底层就是使用迭代器实现的
    - for循环自动处理StopIteration异常
    - 迭代器可以直接用在for循环中
"""
# 1.使用for循环遍历迭代器
fruits = ["苹果", "香蕉", "橘子", "葡萄"]
fruit_iter = iter(fruits)
print("使用for循环遍历迭代器：")
for fruit in fruit_iter:
    print(f"- {fruit}")


# 2.迭代器遍历后变空
colors = ["红", "黄", "蓝"]
color_iter = iter(colors)
print("第一次遍历：")
for c in color_iter:
    print(c, end=" ")
print("\n第二次遍历（已空）：")
for c in color_iter:
    print(c, end=" ")
print("\n第二次没有输出任何内容")


# 3.for循环直接遍历可迭代对象
print("\n直接遍历列表：")
for num in [1, 2, 3, 4, 5]:
    print(num, end=" ")
print("\n\n直接遍历range：")
for i in range(5):
    print(i, end=" ")
print()
```
