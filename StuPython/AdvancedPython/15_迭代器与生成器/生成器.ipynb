```python
"""
生成器函数基础
    - 使用yield关键字的函数就是生成器函数
    - 调用生成器函数返回生成器对象，不会立即执行函数体
    - yield会暂停函数执行并返回值
    - 下次调用next()时从暂停处继续执行
"""
# 1.最简单的生成器函数
def simple_gen():
    print("第一次调用")
    yield 10
    print("第二次调用")
    yield 20
    print("第三次调用")
    yield 30
# 调用生成器函数返回生成器对象
gen = simple_gen()
print(f"生成器对象：{gen}")
print(f"生成器类型：{type(gen)}") # <class 'generator'>
# 使用next()逐步执行
print(f"next()返回：{next(gen)}")
print(f"next()返回：{next(gen)}")
print(f"next()返回：{next(gen)}")


# 2.验证生成器不立即执行
def lazy_gen():
    print("这条消息只有在next()调用时才会打印")
    yield 100
print("创建生成器时不会执行函数体")
gen2 = lazy_gen()
print("调用next()时才开始执行")
print(next(gen2))
```

```python
"""
yield暂停特性
    - yield会暂停函数，保存当前状态
    - 下次next()调用时从上一次yield处继续
    - 可以多次yield产生多个值
"""
# 1.演示yield的暂停和恢复
def counter():
    print("开始执行")
    yield 1
    print("恢复执行，准备返回2")
    yield 2
    print("恢复执行，准备返回3")
    yield 3
    print("函数结束")
print("=== yield暂停演示 ===")
c = counter()
print(f"第1次next：{next(c)}")
print(f"第2次next：{next(c)}")
print(f"第3次next：{next(c)}")


# 2.yield之间的代码都会执行
def show_process():
    print("步骤1完成")
    yield "状态A"
    print("步骤2完成")
    yield "状态B"
    print("步骤3完成")
    yield "状态C"
    print("全部完成")
print("\n=== 带步骤提示的生成器 ===")
p = show_process()
print(f"返回：{next(p)}")
print(f"返回：{next(p)}")
print(f"返回：{next(p)}")
```

```python
"""
while循环中的yield
    - 可以在while循环中使用yield
    - 配合条件控制可以生成有限序列
    - 可以实现更复杂的生成逻辑
"""
# 1.使用while循环生成有限序列
def count_up_to(limit):
    """生成从1到limit的数字"""
    i = 1
    while i <= limit:
        yield i
        i += 1
print("1到5：")
for num in count_up_to(5):
    print(num, end=" ")
print()


# 2.使用break终止生成
def generate_until_5():
    """生成数字，遇到5就停止"""
    i = 1
    while True:
        if i == 5:
            break
        yield i
        i += 1
print("生成直到5（不包括5）：")
for num in generate_until_5():
    print(num, end=" ")
print()


# 3.带条件的生成器
def even_numbers(limit):
    """生成不超过limit的偶数"""
    i = 2
    while i <= limit:
        yield i
        i += 2
print("\n10以内的偶数：")
for num in even_numbers(10):
    print(num, end=" ")
print()
```
