```python
"""
带参数的函数
    - 形参：定义函数时的参数
    - 实参：调用函数时传入的值
    - 参数按顺序传递给形参
"""
# 1.两个参数的加法函数
def add(a, b):
    """计算两个数的和"""
    result = a + b
    print(f"{a} + {b} = {result}")
add(10, 20)
add(100, 200)


# 2.计算矩形面积
def rectangle_area(length, width):
    """计算矩形面积"""
    area = length * width
    print(f"长{length}、宽{width}的矩形面积是：{area}")
rectangle_area(5, 3)
rectangle_area(10, 4)


# 3.显示个人信息
def show_info(name, age, city):
    """显示个人信息"""
    print(f"姓名：{name}，年龄：{age}，城市：{city}")
show_info("张三", 18, "北京")
show_info("李四", 20, "上海")
```

```python
"""
关键字参数
    - 调用函数时指定形参名=值
    - 可以不按照顺序传递参数
    - 关键字参数必须放在位置参数后面
"""
# 1.关键字参数可以不按顺序
def introduce(name, age, job):
    print(f"我叫{name}，今年{age}岁，工作是{job}")
# 位置参数调用（按顺序）
introduce("小明", 25, "工程师")
# 关键字参数调用（不按顺序）
introduce(job="老师", name="小红", age=28)


# 2.混合使用位置参数和关键字参数
def student_info(name, grade, school):
    print(f"{name}，{grade}年级，{school}学校")
# 位置参数必须在关键字参数前面
student_info("王芳", school="一中", grade=3)
# student_info(school="一中", "王芳", 3)  # 错误：关键字参数不能在位置参数前面


# 3.全部使用关键字参数
def calculate(x, y, z):
    print(f"{x} + {y} + {z} = {x + y + z}")
calculate(z=30, x=10, y=20)
```

```python
"""
默认参数
    - 定义函数时给形参设置默认值
    - 有默认值的参数调用时可以不传值
    - 默认参数必须放在没有默认值的参数后面
"""
# 1.带默认参数的函数
def greet(name, greeting="你好"):
    """问候函数，greeting有默认值"""
    print(f"{greeting}，{name}！")
greet("张三")           # 使用默认问候语
greet("李四", "早上好")  # 使用自定义问候语


# 2.带多个默认参数
def create_user(username, age=18, city="北京"):
    print(f"用户名：{username}，年龄：{age}，城市：{city}")
create_user("user1")
create_user("user2", 20)
create_user("user3", 25, "上海")
create_user("user4", city="广州")  # 跳过age参数


# 题目3：默认参数的错误用法（注释演示）
# def wrong_func(a=10, b):  # 错误：默认参数不能在非默认参数前面
#     print(a, b)
```

```python
"""
位置参数和关键字参数混合
    - 调用时位置参数必须在关键字参数前面
    - 同一个参数不能既用位置传参又用关键字传参
"""
# 1.正确混合使用
def show_product(name, price, quantity=1):
    print(f"商品：{name}，单价：{price}，数量：{quantity}")
# 正确：位置参数在前，关键字参数在后
show_product("苹果", 5, quantity=3)
show_product("香蕉", 3)


# 2.错误示例（注释部分会报错）
def test(a, b, c):
    print(a, b, c)
# test(a=10, 20, 30)  # 错误：关键字参数不能在位置参数前面
# test(10, b=20, 10)  # 错误：参数重复
test(10, 20, c=30)     # 正确


# 3.多种调用方式
def order(food, drink="水", dessert=None):
    print(f"主食：{food}，饮料：{drink}，甜点：{dessert}")
order("米饭")                    # 只有位置参数
order("面条", "可乐")             # 两个位置参数
order("饺子", dessert="蛋糕")     # 混合使用
order(food="汉堡", drink="雪碧")  # 全部关键字
```
