```python
"""
*args可变位置参数
    - *args接收多余的位置参数
    - args是元组类型
    - 可以接收任意数量的位置参数
"""
# 1.计算任意数量的数字之和
def sum_all(*args):
    """计算所有传入数字的和"""
    total = 0
    for num in args:
        total += num
    print(f"传入的参数：{args}")
    print(f"总和：{total}")
sum_all(1, 2, 3)
sum_all(10, 20, 30, 40, 50)
sum_all()  # 可以不传参数


# 2.打印所有参数
def print_items(*items):
    """打印所有传入的项目"""
    print(f"共{len(items)}个项目：")
    for i, item in enumerate(items, 1): # enumerate(可迭代对象, start=1)--返回索引（从1开始）和元素
        print(f"  {i}. {item}")
print_items("苹果", "香蕉", "橘子")
print_items("Python", "Java", "C++", "Go")


# 3.*args配合普通参数
def introduce(prefix, *names):
    """前缀加上所有名字"""
    for name in names:
        print(f"{prefix}：{name}")
introduce("同学", "张三", "李四", "王五")
```

```python
"""
**kwargs可变关键字参数
    - **kwargs接收多余的关键字参数
    - kwargs是字典类型
    - 可以接收任意数量的关键字参数
"""
# 1.收集用户信息
def collect_info(**kwargs):
    """收集并打印所有用户信息"""
    print("用户信息：")
    for key, value in kwargs.items():
        print(f"  {key}: {value}")
collect_info(name="张三", age=18, city="北京")
collect_info(name="李四", email="lisi@example.com", phone="123456")


# 2.创建配置字典
def create_config(**settings):
    """创建配置字典并返回"""
    print(f"配置项：{settings}")
    return settings
config = create_config(theme="dark", font_size=14, language="zh")
print(f"返回的配置：{config}")


# 3.**kwargs配合普通参数
def send_message(to, subject, **options):
    """发送消息，可以附加选项"""
    print(f"发送给：{to}")
    print(f"主题：{subject}")
    if options:
        print("附加选项：")
        for k, v in options.items():
            print(f"  {k}={v}")
send_message("admin@example.com", "测试邮件", cc="user@example.com", priority="high")
```

```python
"""
*args和**kwargs同时使用
    - 顺序：普通参数 -> *args -> **kwargs
    - *args接收位置参数
    - **kwargs接收关键字参数
"""
# 1.接收任意参数
def universal_func(*args, **kwargs):
    """接收任意类型的参数"""
    print(f"位置参数：{args}")
    print(f"关键字参数：{kwargs}")
    print("-" * 30)
universal_func(1, 2, 3, name="张三", age=18)
universal_func("hello", "world", a=10, b=20, c=30)
universal_func()


# 2.包装函数
# 定义装饰器函数 logger
def logger(func):
    """记录函数调用的包装器"""
    def wrapper(*args, **kwargs):
        print(f"调用函数：{func.__name__}")
        print(f"参数：args={args}, kwargs={kwargs}")
        result = func(*args, **kwargs)
        print(f"返回值：{result}")
        return result
    return wrapper
@logger # 装饰器
# 先定义原始的 add 函数
def add(a, b):  # def logger(add)(a, b) --> def wapper(a, b)
    return a + b
# 然后执行：add = logger(add)
add(3, 5)   # 实际调用的是 wrapper 函数 -- wrapper(3, 5)
add(a=10, b=20)
"""
add(3, 5) 代码实现过程
1. 定义装饰器函数 logger 此时返回的只是wrapper函数对象 没有调用，只返回
2. 使用 @logger 装饰 add -- 先定义原始的 add 函数，然后执行：add = logger(add)
    此时 def add(a, b) 等价于 def logger(add)(a, b)
3. 先执行 logger(add)，注意此时func=add ，返回wrapper函数对象
    因此 def logger(add)(a, b) ==> def wrapper(a, b)
    接着调用 wrapper 函数
4. def wrapper(*args, **kwargs) ==> wrapper((3, 5), {})
    func 是 wrapper 外部变量，指向原始的 add 函数
    每个函数对象都有一个 __name__ 属性，存储函数的名字
    所以 func.__name__ 就是原始函数的名称 “add”
    输出：调用函数：add
    因为 args = (3,5), kwargs = {} -- 没有关键字参数，所以
    输出：参数：args=(3, 5), kwargs={}
    result = func(*args, **kwargs) 这里等价于 result = add(*args, **kwargs)
5. 调用原始函数 add ==> add(3, 5) ==> 返回 8
6. 输出：返回值：8

"""

# 3.函数参数转发
def original_func(x, y, z=0, name="default"):
    print(f"x={x}, y={y}, z={z}, name={name}")
def caller(*args, **kwargs):
    """转发参数给原函数"""
    print("转发前：", args, kwargs) # 转发前： (1, 2) {}
    original_func(*args, **kwargs)
caller(1, 2)    # x=1, y=2, z=0, name=default
caller(1, 2, 3)     # x=1, y=2, z=3, name=default
caller(1, 2, name="自定义")    # x=1, y=2, z=0, name=自定义
```
