```python
"""
全局变量和局部变量
    - 全局变量：在函数外定义，整个文件可访问
    - 局部变量：在函数内定义，只在函数内有效
    - 函数内部优先使用局部变量
"""
# 1.局部变量
def my_func():
    x = 10  # 局部变量
    print(f"函数内部x：{x}")
my_func()
# print(x)  # 错误：函数外部无法访问局部变量


# 2.全局变量
global_var = 100  # 全局变量
def show_global():
    print(f"函数内部访问全局变量：{global_var}")
show_global()
print(f"函数外部访问全局变量：{global_var}")


# 3.局部变量和全局变量同名
name = "全局变量"
def test():
    name = "局部变量"  # 创建新的局部变量，不影响全局
    print(f"函数内部：{name}")
test()
print(f"函数外部：{name}")


# 4.函数内部修改全局变量需要global
count = 0
def increment():
    global count  # 声明要修改全局变量
    count += 1
increment()
increment()
print(f"修改后的全局变量：{count}")
```

```python
"""
global关键字
    - 在函数内部修改全局变量
    - 声明变量是全局的，不是局部变量
    - 在函数内改变全局变量的值
"""
# 1.使用global修改变量
score = 0
def add_score(points):
    global score
    score += points
    print(f"增加{points}分，当前总分：{score}")
add_score(10)
add_score(20)
add_score(5)
print(f"最终分数：{score}")


# 2.global在嵌套函数中
message = "原始消息"
def outer():
    def inner():
        global message  # 修改的是全局变量
        message = "inner修改的消息"
    inner()
    print(f"outer中查看：{message}")  # 仍然是原始消息
outer()
print(f"全局查看：{message}")


# 3.不使用global无法修改
counter = 0
def wrong_increment():
    # counter += 1  # 取消注释会报错：局部变量未定义
    print("直接修改会报错，需要使用global")
# 正确方式
def correct_increment():
    global counter
    counter += 1
correct_increment()
print(f"正确修改后：{counter}")
```

```python
"""
nonlocal关键字
    - 在嵌套函数中修改外层函数的变量
    - 不能用于全局变量
    - 用于闭包场景
"""
# 1.使用nonlocal修改外层变量
def outer():
    x = 10
    print(f"outer开始时x={x}")
    def inner():
        nonlocal x  # 声明修改外层函数的x
        x = 20
        print(f"inner中修改x={x}")
    inner()
    print(f"outer结束后x={x}")
outer()


# 2.不使用nonlocal无法修改
def outer2():
    y = 100
    def inner2():
        # y = 200  # 这会创建新的局部变量
        print(f"inner2中y={y}")  # 只能读取，不能修改
    inner2()
    print(f"outer2中y={y}")
outer2()


# 3.多层嵌套
def level1():
    a = 1
    print(f"level1初始：a={a}")
    def level2():
        nonlocal a  # 修改level1的a
        a = 2
        def level3():
            nonlocal a  # 仍然修改level1的a
            a = 3
        level3()
        print(f"level2中：a={a}")
    level2()
    print(f"level1结束：a={a}")
level1()
```
