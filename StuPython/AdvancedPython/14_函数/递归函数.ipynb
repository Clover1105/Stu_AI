```python
"""
递归函数基础
    - 函数调用自身
    - 必须有终止条件
    - 避免无限递归
"""
# 1.递归计算阶乘
def factorial(n):
    """计算n的阶乘"""
    # 终止条件
    if n == 0 or n == 1:
        return 1
    # 递归调用
    return n * factorial(n - 1)
print(f"5! = {factorial(5)}")
print(f"10! = {factorial(10)}")


# 2.递归求和1到n
def sum_to_n(n):
    """计算1+2+...+n"""
    if n == 1:
        return 1
    return n + sum_to_n(n - 1)
print(f"1到10的和：{sum_to_n(10)}")
print(f"1到100的和：{sum_to_n(100)}")


# 3.递归打印倒计时
def countdown(n):
    """倒计时"""
    if n <= 0:
        print("发射！")
        return
    print(n)
    countdown(n - 1)
countdown(5)


# 4.递归的栈深度（演示递归深度限制）
def deep_recursion(n):
    print(f"深度：{n}")
    deep_recursion(n + 1)
# deep_recursion(1)  # 会达到递归深度限制，默认1000
print(f"Python默认递归深度限制约1000")
```

```python
"""
递归应用-斐波那契数列
    - 斐波那契：1,1,2,3,5,8,13...
    - 每个数是前两个数之和
    - 递归实现直观但效率低
"""
# 1.递归实现斐波那契
def fibonacci(n):
    """返回第n个斐波那契数（从1开始）"""
    if n == 1 or n == 2:
        return 1
    return fibonacci(n - 1) + fibonacci(n - 2)
print("前10个斐波那契数：")
for i in range(1, 11):
    print(f"F({i}) = {fibonacci(i)}")


# 2.打印斐波那契数列
def print_fibonacci(n):
    """打印前n个斐波那契数"""
    for i in range(1, n + 1):
        print(fibonacci(i), end=" ")
    print()
print("前15个斐波那契数：")
print_fibonacci(15)


# 3.效率对比（演示递归效率低）
import time
def fibonacci_slow(n):
    if n <= 2:
        return 1
    return fibonacci_slow(n - 1) + fibonacci_slow(n - 2)
start = time.time()
result = fibonacci_slow(35)
end = time.time()
print(f"递归计算F(35)={result}，耗时：{end-start:.2f}秒")
print("提示：F(40)以上会很慢，因为重复计算太多")
```
