```python
"""
FileNotFoundError异常处理
    - try-except捕获文件不存在异常
    - 程序不会崩溃，可以继续执行
    - else子句：无异常时执行
    - finally子句：无论是否异常都执行
"""

filename = "不存在的文件.txt"

print("=== 尝试打开不存在的文件 ===")
try:
    f = open(filename, "r", encoding="utf-8")
    content = f.read()
    print(content)
except FileNotFoundError as e:
    print(f"捕获到FileNotFoundError：{e}")
    print("文件不存在，程序不会崩溃")
else:
    f.close()
    print("文件读取成功")
finally:
    print("无论是否异常，都会执行这行代码")

print("\n程序继续正常执行")
```

```python
"""
ValueError异常处理
    - 值错误：参数类型正确但值不合适
    - 常见于类型转换、文件操作等
    - 使用except捕获并处理
"""

def read_number_from_file(filename):
    """从文件读取数字"""
    try:
        with open(filename, "r", encoding="utf-8") as f:
            content = f.read().strip()
            number = int(content)
            print(f"读取到的数字是：{number}")
    except FileNotFoundError:
        print(f"文件 {filename} 不存在")
    except ValueError as e:
        print(f"值错误：{e}")
        print("文件内容不是有效的数字")

# 创建测试文件
with open("number.txt", "w", encoding="utf-8") as f:
    f.write("abc")  # 写入非数字内容

read_number_from_file("number.txt")

# 创建有效文件
with open("valid_number.txt", "w", encoding="utf-8") as f:
    f.write("100")

read_number_from_file("valid_number.txt")
```

```python
"""
PermissionError异常处理
    - 权限错误：没有操作文件的权限
    - 文件被占用时也可能触发
    - 提示用户检查权限
"""

import os

def safe_write_file(filename, content):
    """安全地写入文件"""
    try:
        with open(filename, "w", encoding="utf-8") as f:
            f.write(content)
        print(f"成功写入文件：{filename}")
    except PermissionError as e:
        print(f"权限错误：{e}")
        print("请检查文件是否被其他程序占用或是否有写入权限")

# 正常写入
safe_write_file("test_permission.txt", "测试内容")

# 演示：如果文件被设置为只读会触发PermissionError
# 在Windows中可以先创建只读文件测试
# os.chmod("test_permission.txt", 0o444)  # 设置为只读
# safe_write_file("test_permission.txt", "尝试写入只读文件")
```

```python
"""
异常处理完整结构
    - try: 尝试执行的代码
    - except: 捕获特定异常
    - else: 没有异常时执行
    - finally: 无论是否异常都执行
    - raise: 主动抛出异常
"""

def process_file(filename):
    """处理文件的完整异常处理示例"""
    f = None
    try:
        f = open(filename, "r", encoding="utf-8")
        content = f.read()
        print(f"文件内容长度：{len(content)}字符")

    except FileNotFoundError:
        print(f"文件 {filename} 不存在，将创建新文件")
        f = open(filename, "w", encoding="utf-8")
        f.write("这是新创建的文件")

    except PermissionError:
        print(f"没有权限读取文件 {filename}")

    else:
        print("文件读取成功，没有发生异常")

    finally:
        if f:
            f.close()
            print("文件已关闭")
        print("finally块总是执行\n")

# 测试
process_file("existing.txt")
process_file("new_file.txt")
process_file("不存在的文件.txt")

# 主动抛出异常
def validate_age(age):
    if age < 0:
        raise ValueError("年龄不能为负数")
    print(f"年龄：{age}")

try:
    validate_age(-5)
except ValueError as e:
    print(f"捕获到主动抛出的异常：{e}")
```
