Python中的魔术方法（Magic Methods）是一种特殊的方法，它们以双下划线开头和结尾，例如`__init__`，`__str__`，`__add__`等。
这些方法允许您自定义类的行为，以便与内置Python功能（如+运算符、迭代、字符串表示等）交互。

以下是一些常用的Python魔术方法：

    1. ** `__init__(self, ...)`: 初始化对象，通常用于设置对象的属性。
    2. `__str__(self)`: 定义对象的字符串表示形式，可通过`str(object)`或`print(object)`调用。例如，您可以返回一个字符串，描述对象的属性。
    3. `__repr__(self)`: 定义对象的“官方”字符串表示形式，通常用于调试。可通过`repr(object)`调用。
    4. ** `__len__(self)`: 定义对象的长度，可通过`len(object)`调用。通常在自定义容器类中使用。
    5. ** `__getitem__(self, key)`: 定义对象的索引操作，使对象可被像列表或字典一样索引。例如，`object[key]`。
    6. ** `__setitem__(self, key, value)`: 定义对象的赋值操作，使对象可像列表或字典一样赋值。例如，`object[key] = value`。
    7. `__delitem__(self, key)`: 定义对象的删除操作，使对象可像列表或字典一样删除元素。例如，`del object[key]`。
    8. `__iter__(self)`: 定义迭代器，使对象可迭代，可用于`for`循环。
    9. `__next__(self)`: 定义迭代器的下一个元素，通常与`__iter__`一起使用。
    10. `__add__(self, other)`: 定义对象相加的行为，使对象可以使用`+`运算符相加。例如，`object1 + object2`。
    11. `__sub__(self, other)`: 定义对象相减的行为，使对象可以使用`-`运算符相减。
    12. `__eq__(self, other)`: 定义对象相等性的行为，使对象可以使用`==`运算符比较。
    13. `__lt__(self, other)`: 定义对象小于其他对象的行为，使对象可以使用`<`运算符比较。
    14. `__gt__(self, other)`: 定义对象大于其他对象的行为，使对象可以使用`>`运算符比较。
    15. ** `__call__`: 用对象当作函数调用时 自动调用，参数为：第一个参数（创建的对象） 和 后面的参数（传入的参数）


```python
"""`__add__(self, other)`: 定义对象相加的行为，使对象可以使用`+`运算符相加。"""
# 没有__add__时对象不能相加
class Student:
    age = 5
s1 = Student()
s2 = Student()
s1.age = 10
s2.age = 20
try:
    result = s1 + s2
    print(result)
except TypeError:
    print("两个Student对象不能直接相加")



# 添加__add__方法使对象可以相加
class Student:
    age = 5
    def __add__(self, other):
        """定义对象相加的行为"""
        return self.age + other.age
s1 = Student()
s2 = Student()
s1.age = 10
s2.age = 20
result = s1 + s2
print(f"s1 + s2 = {result}")



# 自定义相加规则
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def __add__(self, other):
        """两个点相加：x+x, y+y"""
        return Point(self.x + other.x, self.y + other.y)
    def __str__(self):
        return f"({self.x}, {self.y})"
p1 = Point(1, 2)
p2 = Point(3, 4)
p3 = p1 + p2
print(f"{p1} + {p2} = {p3}")
```

```python
"""`__init__(self, ...)`: 初始化对象，通常用于设置对象的属性。"""
# 基本的__init__方法
class Student:
    life = 1

    def __init__(self, name, age):
        print("__init__方法被调用了")
        self.name = name
        self.age = age
# 创建对象时传入参数
s1 = Student("小王", 18)
print(f"姓名：{s1.name}, 年龄：{s1.age}")
s2 = Student("小李", 20)
print(f"姓名：{s2.name}, 年龄：{s2.age}")


# 带默认参数的__init__
class Student:
    life = 1
    def __init__(self, name, age, life=5):
        print("__init__方法被调用了")
        self.name = name
        self.age = age
        self.life = life
s1 = Student("小王", 18, 10)
print(f"{s1.name}：年龄{s1.age}，生命值{s1.life}")
s2 = Student("小李", 20)  # 使用默认life值
print(f"{s2.name}：年龄{s2.age}，生命值{s2.life}")


# __init__不修改类属性
print(f"\n类属性life仍然是：{Student.life}")
```

```python
# 完善的学生类
class Student:
    school = "华清远见"
    def __init__(self, name, age, money=0):
        """初始化学生对象"""
        self.name = name
        self.age = age
        self.money = money
    def __add__(self, other):
        """两个学生相加：合并他们的钱"""
        return Student(
            f"{self.name}&{other.name}",
            max(self.age, other.age),
            self.money + other.money
        )
    def __str__(self):
        """打印学生信息"""
        return f"学生({self.name}, {self.age}岁, {self.money}元)"
# 创建学生
s1 = Student("小王", 18, 100)
s2 = Student("小李", 20, 200)
print(s1)
print(s2)
# 学生相加
s3 = s1 + s2
print(f"\n{s1} + {s2} = {s3}")



# 银行账户类
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance
    def __add__(self, other):
        """账户相加：合并余额"""
        return BankAccount(
            f"{self.owner}&{other.owner}",
            self.balance + other.balance
        )
    def __str__(self):
        return f"{self.owner}的账户，余额：{self.balance}元"
acc1 = BankAccount("张三", 500)
acc2 = BankAccount("李四", 300)
acc3 = acc1 + acc2
print(f"\n{acc1}")
print(f"{acc2}")
print(f"合并后：{acc3}")
```

```python
"""`__call__(self, ...)`: 用对象当作函数调用时 自动调用，参数为：第一个参数（创建的对象） 和 后面的参数（传入的参数）"""
# 基本__call__方法
class MyModel:
    def __call__(self):
        print("把对象当做函数调用时，其实就是调用这个方法")
model = MyModel()
model()  # 对象当作函数调用


# 带参数的__call__
class Multiplier:
    def __init__(self, factor):
        self.factor = factor
    def __call__(self, x):
        return x * self.factor
double = Multiplier(2)
triple = Multiplier(3)
print(f"5的2倍：{double(5)}")
print(f"5的3倍：{triple(5)}")
```

```python
# 取值
"""`__getitem__(self, key)`: 定义对象的索引操作，使对象可被像列表或字典一样索引。例如，`object[key]`。"""
# 基本__getitem__实现
class A:
    arr = [10, 20, 30, 40, 50]
    def __getitem__(self, item):
        print(f"__getitem__被调用，传入的索引：{item}")
        return self.arr[item]
a1 = A()
res = a1[3]
print(f"索引3的值：{res}")


# 返回网页标题和网址
class WebPages:
    def __init__(self):
        self.pages = [
            {"url": "http://www.baidu.com", "title": "百度"},
            {"url": "http://www.hqyj.com", "title": "华清远见"},
            {"url": "http://www.taobao.com", "title": "淘宝"}
        ]
    def __getitem__(self, item):
        print(f"__getitem__被调用，传入的索引：{item}")
        return self.pages[item]["title"], self.pages[item]["url"]
pages = WebPages()
title, url = pages[2]
print(f"标题：{title}，网址：{url}")


# 切片支持
class SliceList:
    def __init__(self):
        self.arr = [10, 20, 30, 40, 50, 8, 9, 10]
    def __getitem__(self, item):
        # 支持切片操作，每次返回两个元素
        return self.arr[item * 2:item * 2 + 2]
sl = SliceList()
arr = sl[1]  # 返回索引2和3的元素
print(f"sl[1]返回：{arr}")
```

```python
# 存值
"""`__setitem__(self, key, value)`: 定义对象的索引赋值操作，使对象可被像列表或字典一样索引赋值。例如，`object[key] = value`。"""
# 基本__setitem__实现
class A:
    arr = []
    def __setitem__(self, key, value):
        print(f"__setitem__被调用，传入的索引：{key}，传入的值：{value}")
        self.arr.append((key, value))
    def __getitem__(self, item):
        for key, value in self.arr:
            if key == item:
                return value
        return None
a1 = A()
a1[10] = 100    # 相当于调用了__setitem__(a1, 10, 100)
a1[20] = 200
print(f"键10的值：{a1[10]}")
print(f"键20的值：{a1[20]}")
```

```python
"""
__del__ 魔术方法
    - 当对象被销毁时自动调用
    - 语法：def __del__(self)
    - 使用del语句删除对象时触发
    - 用于清理资源（关闭文件、释放连接等）
"""

# 基本__del__方法
class A:
    def __del__(self):
        print("对象被销毁了")
a1 = A()
del a1  # 触发__del__
print("程序继续执行...")


# 带资源清理的类
class FileHandler:
    def __init__(self, filename):
        self.filename = filename
        print(f"打开文件：{filename}")
    def __del__(self):
        print(f"关闭文件：{self.filename}")
fh = FileHandler("test.txt")
del fh


# 题目3：多个对象的销毁顺序
class Student:
    def __init__(self, name):
        self.name = name
    def __del__(self):
        print(f"学生{self.name}被销毁")
s1 = Student("张三")
s2 = Student("李四")
del s1
print("删除s1后")
del s2
print("程序结束")
```

```python
"""
__sub__ 魔术方法
    - 当对象使用-运算符时自动调用
    - 语法：def __sub__(self, other)
    - 定义两个对象相减的行为
"""

# 基本__sub__实现
class A:
    age = 10
    def __sub__(self, other):
        return self.age - other.age
a1 = A()
a1.age = 100
a2 = A()
a2.age = 200
res = a1 - a2   # a1-->self, a2-->other
print(f"a1 - a2 = {res}")


# 题目2：计算时间差
class Time:
    def __init__(self, hours, minutes):
        self.total_minutes = hours * 60 + minutes
    def __sub__(self, other):
        diff = abs(self.total_minutes - other.total_minutes)
        h = diff // 60
        m = diff % 60
        return Time(h, m)
    def __str__(self):
        return f"{self.total_minutes // 60}小时{self.total_minutes % 60}分钟"
t1 = Time(10, 30)
t2 = Time(8, 15)
diff = t1 - t2
print(f"时间差：{diff}")


# 题目3：金额减法
class Money:
    def __init__(self, amount):
        self.amount = amount
    def __sub__(self, other):
        return Money(self.amount - other.amount)
    def __str__(self):
        return f"{self.amount}元"
wallet = Money(500)
cost = Money(128)
remain = wallet - cost
print(f"剩余：{remain}")
```

```python
"""
__iter__ 和 __next__ 魔术方法
    - __iter__：返回迭代器对象，使对象可迭代，固定返回self
    - __next__：每次迭代返回下一个值，结束时抛出StopIteration
    - 实现这两个方法可以让对象用于for循环
"""

# 基本迭代器实现
class A:
    def __iter__(self):
        print("__iter__被调用")
        return self  # 固定返回self

    def __next__(self):
        return 100  # 每次都返回100
a1 = A()
a2 = iter(a1)
print(f"a1和a2是同一个对象：{a1 is a2}")
print(f"next(a1)：{next(a1)}")
print(f"next(a1)：{next(a1)}")


# 有限次数的迭代器
class BigRange:
    def __init__(self, end):
        self.end = end
        self.n = 0
    def __iter__(self):
        print("__iter__被调用")
        return self
    def __next__(self):
        m = self.n
        self.n += 1
        if self.n > self.end:
            raise StopIteration  # 迭代结束必须抛出此异常
        else:
            return m ** 2  # 返回平方数
print("\n平方数迭代器：")
a1 = BigRange(5)
for i in a1:
    print(i, end=" ")
print()


# 斐波那契迭代器
class Fibonacci:
    def __init__(self, count):
        self.count = count
        self.current = 0
        self.a, self.b = 0, 1
    def __iter__(self):
        return self
    def __next__(self):
        if self.current >= self.count:
            raise StopIteration
        self.current += 1
        result = self.b
        self.a, self.b = self.b, self.a + self.b
        return result
print("\n斐波那契数列：")
for num in Fibonacci(10):
    print(num, end=" ")
print()
```
